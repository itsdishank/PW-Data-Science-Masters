### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Scatter plot: A scatter plot is used to visualize the relationship between two numerical variables. It can help identify any patterns or trends in the data, and is useful for identifying outliers.

2. Line plot: A line plot is used to visualize the relationship between two variables where one variable is continuous and the other variable is categorical. It can be used to display trends over time, or to compare multiple groups.

3. Bar plot: A bar plot is used to visualize the relationship between a categorical variable and a numerical variable. It can be used to compare the values of different categories, or to show the distribution of a single variable.

4. Histogram: A histogram is used to visualize the distribution of a numerical variable. It can help identify the central tendency of the data, as well as any skewness or outliers.

5. Heatmap: A heatmap is used to visualize the relationship between two variables in a dataset. It can be used to identify correlations between variables or to visualize the distribution of values within a matrix.

### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.boxplot(x="pclass", y="age", data=titanic)
sns.boxplot(x="pclass", y="fare", data=titanic)
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

diamonds = sns.load_dataset("diamonds")

sns.histplot(data=diamonds, x="price", hue="cut")
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns

iris = sns.load_dataset("iris")

sns.pairplot(data=iris, hue="species")
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

flights = sns.load_dataset("flights")

flights_matrix = flights.pivot(index="month", columns="year", values="passengers")

sns.heatmap(data=flights_matrix, cmap="YlGnBu")
```
