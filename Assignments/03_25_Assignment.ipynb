### Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.replace('k','')
df['Size']=df['Size'].replace('Varies with device',np.nan)
df['Size']=df['Size'].replace('1,000+','50000')
df['Size']=df['Size'].astype(float)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Size', y='Rating', data=df)
plt.xlabel('Size')
plt.ylabel('Rating')
plt.title('App Size vs. Rating')
plt.ylim(0, 6)
plt.show()
```

* As we can observe from the Scatterplot, there are more apps of size around 0-20000k 
* All the apps with bigger size has higher rating but the apps with smaller size has higher ratings as well as some amount of lesser ratings

### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df[cols]=df[cols].str.replace(item,'')
```

```python
df['Price']=df['Price'].replace('Everyone','0')
```

```python
df['Price'].unique()
```

```python
df['Installs'].unique()
```

```python
df['Price']=df['Price'].astype('float')
```

```python
average_prices = df.groupby('Type')['Price'].mean()
```

```python
plt.figure(figsize=(8, 6))
average_prices.plot(kind='bar', color='blue')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Price by App Type')
plt.xticks(rotation=0)
plt.show()
```

As we can observe the categories for type Free and 0 have the prices of the apps as 0 and the apps of paid category have the prices more than 0
