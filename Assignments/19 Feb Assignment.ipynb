### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a web framework for Python that provides a simple and efficient way to build web applications. It is a micro-framework, meaning it is minimalistic and does not require particular tools or libraries. Flask is popular among developers because of its flexibility and ease of use. Flask offers several advantages, such as:

* Lightweight: Flask is a minimalistic framework and has a small codebase, making it easy to understand and use.
* Modular: Flask provides a modular design, and developers can choose which libraries or tools to include in their projects.
* Flexible: Flask is highly customizable, and developers can easily customize the application according to their needs.
* Easy to learn: Flask has a straightforward API, and developers can learn it quickly, even if they have limited experience in web development.
* Extensible: Flask is easily extensible, and developers can add new functionalities using Flask extensions.

### Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is a mechanism that maps URLs to functions that handle requests. In Flask, we use app routes to define URLs and their associated functions. When a user sends a request to a URL, Flask uses the associated function to generate a response.

We use app routes to define the URLs that our web application will respond to. By using app routes, we can create different pages or functionalities in our web application. It allows us to define the behavior of our web application when users navigate to specific URLs.

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

Flask provides the url_for() function for building URLs dynamically in a Flask application. It takes the name of the endpoint function as its first argument and any arguments that should be appended to the URL as keyword arguments.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return f"""Hello, {name}!</br>
    <a href="/">Back</a>
    """

@app.route('/blog/<int:id>')
def blog_post(id):
    return f"""This is the blog post #{id}</br>
    <a href="/">Back</a>
    """

@app.route('/')
def index():
    return f"""<h1>url_for() function Demo!</h1>
    <p>Example URLs:</p>
    <ul>
        <li><a href="{url_for('hello', name='Dishank')}">{url_for('hello', name='Dishank')}</a></li>
        <li><a href="{url_for('blog_post', id=5)}">{url_for('blog_post', id=5)}</a></li>
    </ul>
    """

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
