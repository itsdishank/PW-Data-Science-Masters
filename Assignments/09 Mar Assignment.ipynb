### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

The Probability Mass Function (PMF) and Probability Density Function (PDF) are both used to describe the probability distribution of a random variable.

A PMF is used to describe the probability distribution of a discrete random variable. It gives the probability of each possible value that the random variable can take on. The sum of all the probabilities in a PMF is equal to 1.

For example, suppose we have a random variable X representing the number of heads obtained when a fair coin is flipped three times. The possible values of X are 0, 1, 2, and 3. The PMF of X is:

P(X=0) = 1/8
P(X=1) = 3/8
P(X=2) = 3/8
P(X=3) = 1/8

This PMF tells us that the probability of getting zero heads is 1/8, the probability of getting one head is 3/8, the probability of getting two heads is 3/8, and the probability of getting three heads is 1/8.

A PDF, on the other hand, is used to describe the probability distribution of a continuous random variable. It gives the relative likelihood of observing a particular value of the random variable. The area under the PDF curve over a range of values gives the probability of the random variable falling within that range.

### Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

The Cumulative Density Function (CDF) is a function that gives the probability that a random variable X is less than or equal to a certain value x, for all possible values of x.

Formally, the CDF of a random variable X is defined as:

F(x) = P(X <= x)

In other words, the CDF gives the cumulative probability up to a certain value of x.

For example, suppose we have a random variable X representing the number of heads obtained when a fair coin is flipped three times. The possible values of X are 0, 1, 2, and 3. The CDF of X is:

F(0) = P(X <= 0) = 1/8
F(1) = P(X <= 1) = 4/8
F(2) = P(X <= 2) = 7/8
F(3) = P(X <= 3) = 8/8 = 1

This CDF tells us that the probability of getting zero or fewer heads is 1/8, the probability of getting one or fewer heads is 4/8, the probability of getting two or fewer heads is 7/8, and the probability of getting three or fewer heads is 1.

The CDF is often used in probability and statistics to calculate probabilities and percentiles for a given distribution. It is also used to visualize the distribution of a random variable, as it gives a cumulative view of the probabilities of different values. The shape of the CDF can reveal information about the skewness, kurtosis, and other characteristics of the distribution.

The standard deviation determines the width of the distribution, with larger values of σ resulting in wider distributions. As the standard deviation gets smaller, the peak of the curve becomes taller and narrower. As the standard deviation gets larger, the peak of the curve becomes shorter and wider.

### Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

The Bernoulli distribution is a discrete probability distribution that models a single binary event with a probability of success (p) and failure (1-p).

An example of a Bernoulli distribution is flipping a coin, where the probability of getting heads (success) is p and the probability of getting tails (failure) is 1-p.

The probability mass function (PMF) of the Bernoulli distribution is:

P(X = 1) = p
P(X = 0) = 1-p

Where X is a random variable that takes on the value of 1 for success and 0 for failure.

The difference between Bernoulli and Binomial distributions is that Bernoulli distribution models a single binary event, while the binomial distribution models the number of successes in a fixed number (n) of independent Bernoulli trials. The binomial distribution is the sum of n independent and identically distributed Bernoulli random variables.

### Q8: What is the z score? State the importance of the z score.

The z-score, also known as the standard score, is a statistical measure that quantifies the distance between a data point and the mean of a distribution in units of standard deviation. It is expressed as the number of standard deviations a data point is away from the mean.

The formula for calculating the z-score is:

z = (x - μ) / σ

Where x is the data point, μ is the mean of the distribution, and σ is the standard deviation of the distribution.

The z-score is important because it allows us to compare and analyze data that are measured on different scales and in different units. By standardizing the data, we can compare observations from different distributions and determine how unusual or extreme a particular observation is relative to the rest of the data.

In addition, the z-score is used to find the area under the normal distribution curve and calculate probabilities. Standard normal distribution tables or calculators use z-scores to determine the probability of a data point falling within a certain range or above/below a certain threshold.



### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem (CLT) is a fundamental theorem in statistics that states that the distribution of the sample means of a large number of independent and identically distributed (i.i.d.) random variables, regardless of the shape of their underlying distribution, tends to follow a normal distribution as the sample size increases.

In other words, if we take repeated random samples of size n from any population, the distribution of the sample means will be approximately normal, regardless of the shape of the original population distribution, as long as the sample size is sufficiently large.

The significance of the Central Limit Theorem is that it allows us to make inferences about the population mean and standard deviation from a sample mean and standard deviation. It also provides a framework for hypothesis testing and confidence interval estimation. The Central Limit Theorem is widely used in statistical analysis and inference, particularly in fields such as economics, finance, and engineering.

### Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem (CLT) makes several assumptions about the underlying population distribution and the sampling process. The assumptions are as follows:

1. Random Sampling: The samples are selected randomly from the population, and each sample is independent of the other samples.
2. Sample Size: The sample size is large enough, typically n ≥ 30, although this may vary depending on the shape of the population distribution.
3. Finite Population: If the population is infinite, the sample size should be less than 10% of the population size.
4. Independence: Each observation in the sample is independent of each other.
5. Identically Distributed: The sample is drawn from a population with a finite mean and standard deviation.
